```python
# Number of matches in which at least 1 collapse

# First create list of matches by Team and Date (unique on both those keys)
gr_matches_and_collapse = df.groupby(['Team', 'MatchDate'])['MatchId'].unique()

# Then aggregate again by Team and Date
gr_matches_w_collapse = gr_matches_and_collapse.groupby(['Team', 'MatchDate']).count().reset_index()

# Then count the number of matches where each Team suffers at least one collapse
gr_matches_w_collapse.groupby(['Team']).count().reset_index()
```

```python
# Number of matches by year in which at least 1 collapse
gr_collapses_by_team_yr = gr_matches_w_collapse.groupby(
    ['Team', 
     gr_matches_w_collapse['MatchDate'].dt.strftime("%Y").rename('year')]
)['MatchId'].count().reset_index()
```
